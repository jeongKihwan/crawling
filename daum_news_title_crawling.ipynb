```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q=파이썬'

# path = "#clusterResultUL > li:nth-child(2) > div.wrap_cont > div > div > a"
path = "#clusterResultUL > li > div.wrap_cont > div > div > a"
```

```python
html = requests.get(path)
print(html)
print(html.text)
```

```python
soup = BeautifulSoup(html.text, 'html.parser')
res = soup.select(path)
```

```python
for i, t in enumerate(res):
    print(i, ':', t.text)
```
