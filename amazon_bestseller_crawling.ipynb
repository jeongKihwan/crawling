```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
url = 'https://www.amazon.com/bestsellers?id=NSGoole'
```

```python
driver = webdriver.Chrome('./chromedriver')
driver.get(url)
```

```python
driver.find_element_by_link_text('Appliances').click()
```

```python
html = driver.page_source
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
lis = soup.select('ol#zg-ordered-list')
lis = lis[0]
```

```python
rank = lis[0].select('span.a-size-small')[0].text
rank
```

```python
lis = lis.select('li')
```

```python
for li in lis:
    rank = li.select('span.a-size-small')[0].text
    title = li.select('div.p13n-sc-truncate-desktop-type2.p13n-sc-truncated')[0].text
    comments = li.select('a.a-size-small.a-link-normal')[0].text
    price = li.select('a.a-link-normal.a-text-normal')[0].text

    print('----------------')
    print(rank)
    print(title)
    print(comments)
    print(price)
```
