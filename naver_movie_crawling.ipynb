```python
for j in range(1, 11):
    url = 'https://movie.naver.com/movie/point/af/list.nhn?&page={}'.format(j)
    res = requests.get(url)

    soup = BeautifulSoup(res.text, 'html.parser')
    path = 'tbody > tr'
    tags = soup.select(path)

    for i in range(10):
        title = tags[i].select('a.movie.color_b')[0].text
        score = tags[i].select('em')[0].text
        text = tags[i].select('td.title')[0].text.split('\n')[5]

        print()
        print('======',i)
        print('제목:',title)
        print('평점:',score)
        print('후기:',text)
```
