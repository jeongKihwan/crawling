```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
driver = webdriver.Chrome()
driver.get('http://ecos.bok.or.kr/jsp/vis/keystat/#/key')
```

```python
# '국민소득.경기.기업경영' section지표를 출력

elems = driver.find_elements_by_css_selector('a.a-c1-list.ng-binding')
elems
```

```python
for e in elems:
    print(e.text)
```

```python
# '국민소득.경기.기업경영' section지표를 선택하고 상세 페이지에서 상세 정보 출력.
```

```python
eco_index = input("'국민소득.경기.기업경영' 관련 경제지표를 선택하세요..")
```

```python
for e in elems:
    if eco_index == e.text:
        e.click()
        break
```

```python
html = driver.page_source
```

```python
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
tags = soup.select('tr.ng-scope')
len(tags)
```

```python
result = []
for t in tags:
    print(t.text.split('\n')[1:4])
    result.append(t.text.split('\n')[1:4])
```
